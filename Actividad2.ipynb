Al revisar la estructura de nuestro dataset, podemos ver una mezcla de distintos tipos de datos. Tenemos variables puramente numéricas que nos servirán para cálculos, como Age y Fare, que representan la edad y la tarifa pagada. También contamos con SibSp y Parch, que son números discretos que nos indican la cantidad de familiares que viajaban con cada pasajero. Por otro lado, aunque la columna Pclass está registrada como un número entero, en realidad funciona como una variable ordinal, ya que la primera, segunda y tercera clase representan una jerarquía clara. El resto de las columnas son en su mayoría categóricas, como Sex o Embarked, o identificadores únicos como Name y Ticket.

Para este paso vamos a checar las variables numericas, ya que las ordinales no tiene mucho sentido sacar las estadisticas descriptivas, como lo son el sexo, embarked y los identificadores unicos

```python
## Seleccionamos las columnas que son numericas para hacer el analisis de sus medias, medianas y modas
col_num = ['Age', "Fare","SibSp","Parch"]

print(f"Media del dataset:\n {df[col_num].mean()}")
print(f"Mediana del dataset:\n {df[col_num].median()}")
print(f"Desviacion Estandar del dataset:\n {df[col_num].std()}")
```

Al revisar los datos, podemos ver que cada variable se comporta de una forma diferente. Para la Edad, notamos que la distribución es bastante simétrica. Esto lo podemos ver porque su media (29.7) y su mediana (28.0) son valores muy parecidos. La desviación estándar de 14.5 nos indica también que había una gran variedad en las edades de los pasajeros. Por otro lado, la Tarifa es todo lo contrario, es muy asimétrica. Su media de 32.2 es mucho mayor que la mediana de 14.45, lo que nos muestra un claro sesgo a la derecha. Esto suele pasar cuando unos pocos valores muy altos "jalan" el promedio hacia arriba, y la alta desviación estándar (49.7) lo confirma.

Con las variables que cuentan familiares, SibSp y Parch, pasa algo interesante. Lo más importante es que en ambas la mediana es 0. Esto significa que la mayoría de los pasajeros no viajaban con hermanos, cónyuges, padres o hijos. Al tener tantos valores en cero, estas distribuciones también están muy sesgadas. En resumen, mientras la edad se distribuye de una manera más o menos normal, las variables de la tarifa y los familiares muestran datos muy concentrados en un extremo.
