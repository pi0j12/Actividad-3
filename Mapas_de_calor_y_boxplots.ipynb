¿Hay alguna variable que no aporta información?

Sí, la variable CustomerID no aporta información relevante para el análisis, ya que únicamente funciona como un identificador único de cada cliente. No tiene valor explicativo ni predictivo.

Si tuvieras que eliminar variables, ¿cuáles quitarías y por qué?

Se eliminaría CustomerID, porque no influye en la segmentación ni en los patrones de comportamiento.
Las demás variables (Gender, Age, Annual Income (k$), Spending Score (1-100)) sí aportan información útil para describir y clasificar a los clientes.

¿Existen variables que tengan datos extraños?

Al analizar con diagramas de caja y bigotes:

Age: no presenta valores atípicos evidentes.

Annual Income (k$): se detectan 2 valores atípicos (clientes con ingresos muy altos en comparación con el resto).

Spending Score (1-100): no presenta outliers significativos, ya que está acotada entre 1 y 100.

Si comparas las variables, ¿todas están en rangos similares? ¿Crees que esto afecte?

No, los rangos son distintos:

Edad: ~18–70

Ingreso anual: ~15–137

Spending Score: 1–100

Esto sí afecta, porque los algoritmos de clustering basados en distancias (como k-means) son sensibles a la escala. Por lo tanto, es necesario normalizar o estandarizar los datos antes de aplicar estos métodos.

¿Puedes encontrar grupos que se parezcan? ¿Qué grupos son estos?

Sí, aplicando k-means sobre Ingreso anual y Spending Score se encuentran grupos de clientes bien definidos. Por ejemplo:

Clientes de bajos ingresos con bajo gasto.

Clientes de altos ingresos pero bajo gasto.

Clientes de ingresos medios con gasto alto.

Clientes con ingresos y gasto altos (segmento premium).
